```python
def remove_letters(l,s):
    l1=list(s)
    for i in range(len(s)):
        if l1[i] in l:
            l.remove(l1[i])
    return l
            
```

```python
remove_letters(["s", "t", "r", "i", "n", "g", "w"], "string") 
```

```python
remove_letters(["b", "b", "l", "l", "g", "n", "o", "a", "w"], "balloon") 
```

```python
remove_letters(["d", "b", "t", "e", "a", "i"], "edabit")
```

```python
def blocks(n):
    steps=5
    if n == 1:
        steps=5
    else:
        for i in range(2,n+1):
            steps = steps+i+5
    return steps
        
```

```python
blocks(5)
```

```python
blocks(3)
```

```python
blocks(2)
```

```python
blocks(4)
```

```python
def my_sub(n1,n2):
    l1,l2=[],[]
    for i in range(n1):
        l1.append(1)
    for i in range(n2):
        l2.append(1) 
    if len(l1)>len(l2):
        for i in range(len(l2)):
            if l2[i] in l1:
                l1.remove(l2[i])
        return len(l1)
    else:
        for i in range(len(l1)):
            if l1[i] in l2:
                l2.remove(l1[i])
        return len(l2)
                
        
```

```python
my_sub(5, 9)
```

```python
my_sub(10, 30)
```

```python
def add_bill(s):
    l1=[]
    summ=0
    l=s.split(",")
    for i in l:
        if  "d" in i:
            
            l1.append(i)
    for i in range(len(l1)):
        if 'k' in l1[i]:
            l1[i]=l1[i].replace('k',"000")
        if 'd' in l1[i]:
            l1[i]=l1[i].replace('d',"")
    for i in range(len(l1)):
        summ=summ+int(l1[i])
    return summ
```

```python
add_bill("d20,p40,p60,d50")
```

```python
add_bill("p30,d20,p60,d150,p360")
```

```python
add_bill("p30,d2k,p60,d200,p360")
```

```python
def flip_list(in_list):
    if len(in_list) > 0:
        output = [ele[0] for ele in in_list] if isinstance(in_list[0],list) else [[ele] for ele in in_list]
    else:
        output = []
    print(output)
            
```

```python
flip_list([1, 2, 3, 4])
```

```python
flip_list([[5], [6], [9]])
```
