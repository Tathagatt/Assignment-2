```python
def fib_fast(n):
    if n <= 1:
        return n
    else:
        return(fib_fast(n-1) + fib_fast(n-2))

```

```python
fib_fast(5)
```

```python
fib_fast(10)
```

```python
fib_fast(20)
```

```python
def convert_to_hex(s):
    hexa = ""
    for i in range(len(s)):
        ch = s[i]
        l = ord(ch)
        p = hex(l).lstrip("0x").rstrip("L")
        hexa += p
    return hexa     
```

```python
convert_to_hex("hello world")
```

```python
convert_to_hex("big Boi")
```

```python
convert_to_hex("Marty Poppinson")
```

```python
def uncensor(s, vowels):
    l1 = list(s)
    l2 = list(vowels)
    vc = 0
    for i in range(len(l1)):
        if l1[i] == "*":
            l1[i] = l2[vc]
            vc += 1

    result = ""
    for i in l1:
        result += i
    
    return result
```

```python
uncensor("Wh*r* d*d my v*w*ls g*?", "eeioeo")
```

```python
uncensor("abcd", "")
```

```python
uncensor("*PP*RC*S*", "UEAE")
```

```python
def get_domain(ip_address):
    import socket
    result=socket.gethostbyaddr(ip_address)
    return list(result)[0]
```

```python
get_domain("8.8.8.8")
```

```python
get_domain("13.251.106.90")
```

```python
get_domain("8.8.4.4")
```

```python
def fact(n):
    result = 1
    for i in range(1,n+1):
        result = result*i
    return result
        
def fact_of_fact(n):
    result = 1
    for i in range(1,n+1):
        result= result*fact(i)
    return result
```

```python
fact_of_fact(4)
```

```python
fact_of_fact(5)
```

```python
fact_of_fact(6)
```
