```python
def show_the_love(lst):
    m=min(lst)
    s=0
    for i in range(len(lst)):
        if lst[i]!= m:
            s=s+lst[i]*0.25
            lst[i]=lst[i]*0.75
    for i in range(len(lst)):
        if lst[i] == m:
            lst[i]=lst[i]+s
    return lst
```

```python
show_the_love([4, 1, 4])

```

```python
show_the_love([16, 10, 8])
```

```python
show_the_love([2, 100])
```

```python
def pairs(lst):
    lst1 = lst.copy()
    output = []
    while True:
        if len(lst) > 0:
            if len(lst) == 1:
                output.append([lst[0],lst.pop(0)])
            else:
                output.append([lst.pop(0),lst.pop(-1)])
        else:
            break
    print(f'pairs({lst1}) ➞ {output}')
```

```python
pairs([1, 2, 3, 4, 5, 6, 7])
pairs([1, 2, 3, 4, 5, 6])
pairs([5, 9, 8, 1, 2])
pairs([])
```

```python
def add_str_nums(s1,s2):
    s1 = s1 if len(s1) > 0 else "0"
    s2 = s2 if len(s2) > 0 else "0"
    if s1.isdigit() == False or s2.isdigit() == False:
        output = -1
    else: 
        output = int(s1)+int(s2)
    print(output)
        
add_str_nums("4", "5")
add_str_nums("abcdefg", "3")
add_str_nums("1", "")
add_str_nums("1874682736267235927359283579235789257", "32652983572985729")
```

```python
def unmix(s1):
    output = ''
    for ele in range(0,len(s1)-1,2):
        output += s1[ele+1]+s1[ele]
        if (len(s1)%2 != 0 and ele == len(s1)//2 ):
            output += s1[-1]       
    print(output)

unmix("123456")
unmix("hTsii  s aimex dpus rtni.g")
unmix("badce")
```
