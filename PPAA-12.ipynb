```python
def meme_sum(n1,n2):
    ns1=str(n1)
    ns2=str(n2)
    ms=''
    if len(ns1)!=len(ns2):
        if len(ns1)>len(ns2):
            z=len(ns1)-len(ns2)
            ns2="0"*z+ns2
        if len(ns2)>len(ns1):
            z=len(ns2)-len(ns1)
            ns1="0"*z+ns1
    else:
        ns1=str(n1)
        ns2=str(n2)
    for i in range(len(ns1)):
        ms=ms+str(int(ns1[i])+int(ns2[i]))
    return int(ms)
        
            
```

```python
meme_sum(26, 39)
```

```python
meme_sum(122, 81)
```

```python
meme_sum(1222, 30277)
```

```python
def prime(n):
    count=0
    for i in range(1,n):
        if n%i==0:
            count+=1
    if count==1:
        output= True
    else:
        output= False
    return output
        
        
```

```python
def next_prime(n):
    if prime(n):
        return n
    else:
        n=n+1
        while(prime(n)==False):
            n+=1
        return n
            

        
    
```

```python
next_prime(11)
```

```python
next_prime(24)
```

```python
next_prime(12)
```

```python
def ave_spd(t1,s1,s2):
    t1=t1/60
    d=s1*t1
    t2=d/s2
    avg_speed=2*d/(t1+t2)
    return avg_speed
```

```python
ave_spd(18, 20, 60) 
```

```python
ave_spd(30, 10, 30) 
```

```python
ave_spd(30, 8, 24)
```

```python
def fact(n):
    f=1
    if n==1 or n==0:
        return 1
    else:
        f = fact(n-1)*n
    return f
```

```python
def kempner(n):
    n1=1
    while((fact(n1))%n!=0):
        n1=n1+1
    return n1
```

```python
kempner(6)
```

```python
kempner(10)
```

```python
kempner(2)
```
