```python
import numpy as np
def get_frame(c,r,ch):
    if c>2:
        l=[]
        for i in range(r):
            l.append([])
        l[0].append(ch*c)
        l[r-1].append(ch*c)
        for j in range(1,r-1):
            l[j].append(ch+(c-2)*" "+ch)
        for i in l:
            print(i)
    else:
        return "INVALID"
        
```

```python
get_frame(4, 5, "#")
```

```python
get_frame(10, 3,"*")
```

```python
get_frame(2, 5, "0")
```

```python
def boolean_and(l):
    while(len(l)>1):
        l1= [bool(l[0]*l[1])]
        l2=l[2:]
        l=l1+l2
    print(l)
def boolean_or(l):
    while(len(l)>1):
        l1= [bool(l[0]+l[1])]
        l2=l[2:]
        l=l1+l2
    print(l)
def boolean_xor(l):
    while(len(l)>1):
        if l[0]!=l[1]:
            l1= [True]
        else:
            l1= [False]
        l2=l[2:]
        l=l1+l2
    print(l)
```

```python
boolean_and([True, True, False, True])
```

```python
boolean_or([True, True, False, False])
```

```python
boolean_xor([True, True, False, False])
```

```python
import numpy as np
def make_box(n):
    if n>0:
        l=[]
        ch='#'
        for i in range(n):
            l.append(" ")
        l[0]=(ch*n)
        l[n-1]=(ch*n)
        for j in range(1,n-1):
            l[j]=(ch+(n-2)*" "+ch)
        for i in l:
            print(i)
    else:
        return "INVALID"
```

```python
make_box(3)
```

```python
make_box(5)
```

```python
make_box(2)
```

```python
make_box(1)
```

```python
def no_duplicate_letters(s):
    o = None
    for ele in s.split(' '):
        if len(ele) == len(set(ele)):
            o = True
        else:
            o = False
            break
    return o
```

```python
no_duplicate_letters("look before you leap")
```

```python
no_duplicate_letters("Fortune favours the bold.") 
```

```python
no_duplicate_letters("You can lead a horse to water, but you can&#39;t make him drink.")
```

```python
no_duplicate_letters("An apple a day keeps the doctor away.") 
```
