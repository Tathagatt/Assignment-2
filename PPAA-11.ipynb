```python
def filter_primes(l):
    lst=[]
    for i in l:
        count=0
        for j in range(1,i):
            if i%j==0:
                count+=1
        if count==1:
            lst.append(i)
    return lst
        
        
```

```python
filter_primes([7, 9, 3, 9, 10, 11, 27])
```

```python
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70])
```

```python
filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093,
1097])
```

```python
def pop(l):
    l1=[]
    l2=[]
    if len(l)==1:
        return l
    else:
        mid=len(l)//2
        for i in range(mid+1):
            l1.append(i)
        for i in range(mid-1,-1,-1):
            l2.append(i)
        return l1+l2
```

```python
pop([0, 0, 0, 0, 4, 0, 0, 0, 0])
```

```python
pop([0, 0, 0, 3, 0, 0, 0])
```

```python
pop([0, 0, 2, 0, 0])
```

```python
pop([0])
```

```python
def loves_me(n):
    s=''
    for i in range(n):
        if i%2==0 and i!=n-1:
            s=s+"Loves me, "
        if i%2==0 and i==n-1:
            s=s+"LOVES ME "
        if i%2!=0 and i!=n-1:
            s=s+'Loves me not, '
        if i%2!=0 and i==n-1:
            s=s+'LOVES ME NOT '
    return s
            
        
```

```python
loves_me(3)
```

```python
loves_me(6)
```

```python
loves_me(1)
```

```python
def sort_by_letter(s):
    l = []
    output = []
    for ele in s:
        for char in ele:
            if char.isalpha():
                l.append(char)
    l.sort()
    for char in l:
        for ele in s:
            if char in ele:
                output.append(ele)
    return output
```

```python
sort_by_letter(["932c", "832u32", "2344b"])
```

```python
sort_by_letter(["99a", "78b", "c2345", "11d"])
```

```python
sort_by_letter(["572z", "5y5", "304q2"])
```

```python
def cup_swapping(swap_list,intitial_ball_pos):
    ball_position = intitial_ball_pos
    for ele in swap_list:
        if ball_position in ele:
            ball_position = ele.replace(ball_position,'')
    print(ball_position)
            
cup_swapping(["AB", "CA", "AB"],'B')
cup_swapping(["AB", "CA", "AB"],'C')
cup_swapping(["AC", "BC", "AB"],'A')
```
