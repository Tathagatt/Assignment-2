```python
def pentagonal(n):
    result=1
    if n == 1:
        return result
    for i in range(n):
        result = result+pentagonal(1)*5*i
    return result
```

```python
pentagonal(1)
```

```python
pentagonal(2)
```

```python
pentagonal(8)
```

```python
pentagonal(3)
```

```python
def encrypt(s):
    s.lower()
    s1 = ""
    dt = {"a":0,"e":1,"i":2,"o":3,"u":4}
    s=s[::-1]
    for i in s:
        if i in dt:
            i = dt[i]
        s1=s1+str(i)
    return s1+"aca"
        
        
```

```python
encrypt("banana")
```

```python
encrypt("karaca")
```

```python
encrypt("burak")
```

```python
encrypt("alpaca")
```

```python
import datetime
from datetime import date

def has_friday_13(month, year):
    s = datetime.date(year, month, 13)
    return s.strftime("%A") == "Friday"
```

```python
has_friday_13(3, 2020)
```

```python
has_friday_13(10, 2017)
```

```python
has_friday_13(1, 1985)
```

```python
import re
def cookies(s):
    pattern = "bad cookie"
    return len(re.findall(pattern, ",".join(s)))
```

```python
cookies(["bad cookie", "good cookie", "bad cookie", "good cookie", "good cookie"])
```

```python
def pluralize(arr):
    lst = []
    for i in range(len(arr)):
        if arr.count(arr[i])>1:
            lst.append(arr[i]+"s")
        else:
            lst.append(arr[i])
    return set(lst)
        
```

```python
pluralize(["cow", "pig", "cow", "cow"]) 
```

```python
pluralize(["table", "table", "table"]) 

```

```python
pluralize(["chair", "pencil", "arm"])
```
