```python
def playback_duration(t,sp):
    l=t.split(":")
    t_new=''
    h=int(l[0])*3600
    m=int(l[1])*60
    s=int(l[2])
    total=h+m+s
    t1=int(total//sp)
    h1=int(t1//3600)
    r1=int(t1%3600)
    m1=int((r1)//60)
    s1=int(r1%60)
    t_new=str(h1)+":"+str(m1)+":"+str(s1)
    return t_new
    
```

```python
playback_duration("00:30:00", 2)
```

```python
playback_duration("01:20:00", 1.5) 
```

```python
import math

def quad_roots( a, b, c):
    dis = b * b - 4 * a * c
    sqrt_val = math.sqrt(abs(dis))

  
    if dis > 0:
        r1 = (-b + sqrt_val)/(2 * a)
        r11=str(r1).split(".")
        r2 = (-b - sqrt_val)/(2 * a)
        r22=str(r2).split(".")
        if r1>0 and r11[1]=="0"  :
                              print(r1)
        elif r2>0 and r22[1]=="0":
                              print(r2)
        else:
            print("none")
    elif dis == 0:
        r=-b / (2 * a)
        rr=str(r).split(".")
        if r>0 and rr[1]=="0":
                              print(r)
        else:
            print("none")
    else:
        print("none")

```

```python
def pile_of_cubes(n):
    c=math.sqrt(n*4)
    quad_roots(1,1,-c)
    
```

```python
def find_fulcrum(lst):
    ful=0
    for i in range(1,len(lst)):
        s1,s2=sum(lst[:i]),sum(lst[i+1:])
        if s1==s2:
            ful=lst[i]
    return ful

        
    
    
```

```python
find_fulcrum([3, 1, 5, 2, 4, 6, -1])
```

```python
find_fulcrum([1, 2, 4, 9, 10, -10, -9, 3])
```

```python
find_fulcrum([9, 1, 9])
```

```python
find_fulcrum([7, -1, 0, -1, 1, 1, 2, 3])
```

```python
find_fulcrum([8, 8, 8, 8])
```

```python
from collections import Counter
def sock_merchant(lst):
    s=0
    a=Counter(lst)
    for i in set(lst):
        s=s+a[i]//2
    return s

        
```

```python
sock_merchant([10, 20, 20, 10, 10, 30, 50, 10, 20])
```

```python
sock_merchant([50, 20, 30, 90, 30, 20, 50, 20, 90])
```

```python
sock_merchant([])
```
