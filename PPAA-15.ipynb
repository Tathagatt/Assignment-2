```python
def anagram(s,lst):
    s=s.lower()
    l=list(s)
    l1=[]
    flag=None
    for i in lst:
        for j in i:
            if j in l:
                l1.append(j)
            else:
                l1.append(" ")
    if len(l1)==len(s.replace(" ",'')):
        flag=True
    else:
        flag=False
            
    return flag
                
```

```python
anagram("Justin Bieber", ["injures", "ebb", "it"]) 
```

```python
anagram("Natalie Portman", ["ornamental", "pita"]) 
```

```python
anagram("Chris Pratt", ["chirps", "rat"])
```

```python
anagram("Jeff Goldblum", ["jog", "meld", "bluffs"])
```

```python
def leaderboards(lst):
    l1=[]
    l2=[]
    for i in lst:
        i["truescore"]=i["score"]+2*i["reputation"]
        truescore=i["score"]+2*i["reputation"]
        l1.append(truescore)
    l1=sorted(l1,reverse=True)
    for i in l1:
        for j in lst:
            if i in j.values():
                j.pop("truescore")
                l2.append(j)
    return l2
```

```python
leaderboards([
  { "name": "a", "score": 100, "reputation": 20 },
  { "name": "b", "score": 90, "reputation": 40 },
  { "name": "c", "score": 115, "reputation": 30 },
])
```

```python
def hangman(s,l):
    out_string = ''
    for ele in range(len(s)):
        if s[ele].lower() in l:
            out_string += s[ele]
        elif s[ele] in '"! ':
            out_string += s[ele]
        else:
            out_string += '-'
    print(out_string)
            
```

```python
hangman("helicopter", ["o", "e", "s"])
hangman("tree", ["r", "t", "e"])
hangman("Python rules", ["a", "n", "p", "r", "z"])
hangman("He\"s a very naughty boy!", ["e", "a", "y"])
```

```python
def digit_sort(in_list):
    max_len = len(str(max(in_list)))
    output = []
    for item in range(max_len,0,-1):
        temp = []
        for ele in in_list:
            if len(str(ele)) == item:
                temp.append(ele)
        output.extend(sorted(temp))
    print(output)
    
digit_sort([77, 23, 5, 7, 101])
digit_sort([1, 5, 9, 2, 789, 563, 444])
digit_sort([53219, 3772, 564, 32, 1])
```
