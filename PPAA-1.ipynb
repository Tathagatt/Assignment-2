```python
def check_score(lst):
    sum = 0
    dt = {"#" : 5,"O" : 3,"X" : 1,"!" : -1,"!!": -3,"!!!" : -5}
    for i in lst:
        for j in i:
            if j in dt:
                sum = sum + dt[j]
    if sum <=0:
        sum = 0
    return sum
    
```

```python
check_score([
["#", "!"],
["!!", "X"]
])
```

```python
check_score([
["!!!", "O", "!"],
["X", "#", "!!!"],
["!!", "X", "O"]
])
```

```python
def combinations(*args):
    mul = 1
    for i in args:
        mul = mul*i
    return mul
```

```python
combinations(3, 7, 4)
```

```python
combinations(2,3, 4,5)
```

```python
combinations(2,3)
```

```python
def prime(x):
    for i in range(2,x-1):
        if x%i==0:
            return False
        else:
            return True
```

```python
prime(7)
```

```python
prime(56963)
```

```python
prime(5151512515524)
```

```python
prime(2**(64-1))
```

```python
def to_boolean_list(s):
    s.lower()
    alpha = 'abcdefghijklmnopqrstuvwxyz'
    lst = []
    for i in s:
        k = alpha.index(i)+1
        if k%2==0:
            k=0
        else:
            k=1
        lst.append(bool(k))
    return lst
        
        
```

```python
to_boolean_list("deep")
```

```python
to_boolean_list("loves")
```

```python
to_boolean_list("tesh")
```
