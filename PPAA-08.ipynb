```python
vowels=['a','e','i','o','u']
def vowel_links(s):
    s=s.lower()
    l=s.split(" ")
    for i in range(len(l)-1):
        if l[i][-1] in vowels and l[i+1][0] in vowels:
            return "True"
    
    return "False"
```

```python
vowel_links("a very large appliance") 
```

```python
vowel_links("go to edabit") 
```

```python
vowel_links("an open fire") 
```

```python
vowel_links("a sudden applause")
```

```python
def first_before_second(s,s1,s2):
    s=s.lower()
    l=s.split(" ")
    flag=True
    l1=[]
    l2=[]
    for i in range(len(s)):
        if s[i]==s1:
            l1.append(i)
        if s[i]==s2:
            l2.append(i)
    for i in l1:
        for j in l2:
            if i<j:
                flag
            else:
                flag=False
    return flag       
    
    
```

```python
first_before_second("a rabbit jumps joyfully", "a", "j") 
```

```python
first_before_second("knaves knew about waterfalls", "k", "w")
```

```python
first_before_second("happy birthday", "a", "y")
```

```python
first_before_second("precarious kangaroos", "k", "a")
```

```python
def char_at_pos(ls,sp):
    l=[]
    if sp=="even":
        for i in range(len(ls)):
            if i%2!=0:
                l.append(ls[i])
    if sp=="odd":
        for i in range(len(ls)):
            if i%2==0:
                l.append(ls[i])
    return l
        
            
```

```python
char_at_pos([2, 4, 6, 8, 10], "even")
```

```python
char_at_pos("EDABIT", "odd")
```

```python
char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd")
```

```python
def palindrome_type(d):
    s=""
    d1=d
    while (d1 > 0):
        r = d1 % 2;
        d1 = int(d1/ 2);
        s=s+str(r)
    if str(d)==str(d)[::-1] and s==s[::-1]:
        return "Both"
    elif str(d)==str(d)[::-1]:
        return "Decimal Only"
    elif s==s[::-1]:
        return "Binary Only"
    else:
        return "neither"
```
